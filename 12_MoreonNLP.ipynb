##üíØ Before you proceed...
You can write the regular expression pattern as <font color = 'red'> either</font> **RegexpTokenizer("[\w]+"), <font color = 'red'> either</font> **RegexpTokenizer("\w+"), <font color = 'red'> either</font> **RegexpTokenizer(r"\w+")** <font color = 'red'>or</font> **RegexpTokenizer(r"[\w]+")**. <font color = 'blue'> The 'r' before the string denotes a raw string literal, which is often used with regular expressions to avoid issues with escape characters.

The regular expression [\w]+ breaks down as follows:

- the backslash \ is known as an escape character. It tells the regex engine, "Don't treat the next character literally; treat it as a special command."
- **\w** shorthand character class, which represents word characters "a-zA-Z0-9".
- [ ] matches any single character specified in a pair of the square brackets.
- **[\w]**: Match any word character (equivalent to [a-zA-Z0-9_]).
- **+**: Match one or more occurrences of the preceding character class.
- So, **[\w]+** will match one or more consecutive word characters, effectively tokenizing text into words.
>
üÜò If you remove the square brackets and use \w+ (without [ ]), it would still work the same way, as \w+ is a shorthand for "match one or more word characters." The square brackets, in this case, are redundant but can be used for clarity or when you want to match a specific set of characters.
>
üÜò The 'r' before the string (as in r"[\w]+") is optional in this case but is a good practice, especially with regular expressions, to avoid unintended escape character interpretation. It tells Python to treat the string as a "raw string," where backslashes are treated literally and not as escape characters as in \t (a new tab) or \n (a new line). However, \w is not a standard escape sequence, and Python treats \w as any word character, not the letter w.

###**A: How to Fix This (The Workaround)**

To properly process collocations for a dispersion plot, you need to Pre-process the N-grams. Here are the two most common methods:

- Manual Bigram Replacement (Underlining)Before passing your text to the dispersion function, you can replace known collocations with a single token joined by an underscore.

- Transformation: "Captain Ahab": "Captain_Ahab"

- Plotting: You then plot the single token Captain_Ahab.

## üî® üî® Example for Pre-processing the N-grams

###<font color = 'sky blue'> üìå **We need to handle multi-word expressions before the tokenization step to ensure they are treated as a single unit.**

```python
!pip install nltk
import nltk
from nltk.tokenize import word_tokenize

# 1. Download the neccesary resources (e.g., Gutenberg corpus in NLTK)
nltk.download('gutenberg')

# 2. Load the raw text from 'austen-emma.txt'
text = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')

# 3. üíä GLUE PHASE (Optimized)
# üìåüìåüìå We use a DICTIONARY or a LIST of TUPLES to keep it organized. Either a dictionary or a list will work fine.
target_collocations = {
    (r"captain\s+Ahab", "Captain_Ahab"),
    (r"captain\s+Boomer", "Captain_Boomer"),
    (r"father\s+Mapple", "Father_Mapple"),
    (r"white\s+whale", "White_Whale"),
    (r"Moby\s+Dick", "Moby_Dick")
}

#üí£üí£ALERT
# Simple use of a dictionary with TUPLES for replacements did not work out...
#collocations = {
#    "Captain Ahab": "Captain_Ahab",
#    "Captian Boomer": "Captain_Boomer",
#    "Father Mapple": "Father_Mapple",
#    "white whale": "White_Whale",
#    "Moby Dick": "Moby_Dick"
#}

# 4. üíä Chaining the replacements: Start with raw_text, then update 'processed_text'
processed_text = text
for pattern, replacement in target_collocations:
    # Notice we update 'processed_text' using 'processed_text'
    processed_text = re.sub(pattern, replacement, processed_text, flags=re.IGNORECASE)


# 5.  TOKENIZE PHASE
# We include "_" in the regex so phrases such as "Moby_Dick" are kept as one single token.
retokenize = RegexpTokenizer(r"[a-zA-Z0-9_]+")
words = retokenize.tokenize(processed_text)

#üí£üí£ALERT
# We wanted to include an underscore in the RE. So, avoid using r"[\w]+" when running this cell.

# 6. CREATE NLTK TEXT OBJECT
text_object = nltk.Text(words)
```

```python
# Generate a dispersion plot (Î∂ÑÏÇ∞ÎèÑ) for the specified words
text_object.dispersion_plot(["Captain_Ahab","Ishmael", "Queequeg", "Moby_Dick", "Starbuck", "Father_Mapple", "Stubb", "Captain_Boomer"])
```

##**N-gram (mathematical tool)vs. Collocation (a linguistic phenomenon.)**


### <font color = 'blue'> **1. N-grams (The Mathematical Window)**

- An N-gram is simply a contiguous sequence of _n_ items from a given sample of text.
- It is a strictly mechanical grouping based on proximity.
  - Unigram (_n=1_): "Captain", "Ahab", "the", "whale"
  - Bigram (_n=2_): "Captain Ahab", "Ahab stood", "the white", "white whale"
  - Trigram (_n=3_): "The white whale", "Ahab stood on"
  - The Key Characteristic: N-grams include everything. Most bigrams are "garbage" (e.g., "of the", "in a", "it is") because they don't represent a unique concept.


### <font color = 'orange'> **2. Collocations (The Meaningful Connection)**
- A Collocation is a sequence of words that co-occur more often than would be expected by chance.
- They "belong" together, e.g., "Captain Ahab", "Strong coffee", "Decision making", "Pay attention".
- How to "Test"?
    - If you replace one word with a synonym, it sounds wrong. We say "strong coffee," but we don't say "powerful coffee," even though the meaning is similar.

```python
# Install and import nltk for tokenizing words
!pip install nltk
import nltk
from nltk.tokenize import word_tokenize

# This imports the WordCloud class and a parameter called STOPWORDS set from the wordcloud library.
from wordcloud import WordCloud, STOPWORDS

# Download the Gutenberg corpus
nltk.download('gutenberg')

# Load the text of Alice's Adventures in Wonderland by Lewis Carroll in 1865
text = nltk.corpus.gutenberg.raw('carroll-alice.txt')

# Create a set of stopwords
stopwords = set(STOPWORDS) #By passing STOPWORDS into set(), you convert it into a mutable set. A mutable set allows modifications like adding or removing items.

# Generate Word Cloud which includes 'the' but excludes other stopwords
wc1 = WordCloud(stopwords = stopwords).generate(text)
#Left side: The stopwords keyword argument is a parameter for the WordCloud class.
#Right side: The second stopwords refers to the local variable.
#This code is prerequsite for the following code line (stopwords.add('the'))
stopwords.add('the') #Among excluded stop words, include 'the' in the analysis.

# Display the Word Cloud using Matplotlib
plt.imshow(wc1)
```
